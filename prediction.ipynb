# Infrence (predition by trained model)

```python
import pickle
import pandas as pd
import numpy as np
```

## load model, scaler, and pca

to sacle new test data and reduce dimension for prediction by trained model

```python
## load trained model, scaler.pkl, epca.pkl

with open('x_scaler.pkl', 'rb') as f:
  scaler = pickle.load(f)

with open('xgb_model.pkl', 'rb') as file:
  xgb_model = pickle.load(file)

# Load PCA model
with open("pca.pkl", "rb") as f:
    pca = pickle.load(f)
```

## function to predict mycotoxins

```python
def infrence(X, scaler, pca_model, model):
  if 'hsi_id' in X.columns:
      X = X.drop(columns=['hsi_id'])
  if 'vomitoxin_ppb' in X.columns:
      X = X.drop(columns=['vomitoxin_ppb'])
  X_scaled = scaler.transform(X)
  X_pca = pca_model.transform(X_scaled)
  y_pred = model.predict(X_pca)
  return y_pred
```

```python
X = pd.read_csv("test.csv")
y_pred = infrence(X, scaler, pca, xgb_model)
```
